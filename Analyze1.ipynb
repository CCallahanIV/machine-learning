```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('boston_housing_data.csv', 'r') as f:
    data = f.readlines()
```

```python
for i in range(len(data)):
    data[i] = ",".join(data[i].split())
data = "\n".join(data)
f2 = open('formatted_data.csv', 'w')
f2.write(data)
f2.close()
```

```python
f_data = pd.read_csv('formatted_data.csv', names=('CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'))
```

This dataset was taken from the StatLib library which is maintained at Carnegie Mellon University and concerns housing values in suburbs of Boston.

```python
f_data
```

```python
df = pd.DataFrame(f_data)
```

```python
'Median Home Value: $' + str(df["MEDV"].median() *1000)
```

```python
'Average Home Value: $' + str(df["MEDV"].mean() * 1000)
```

```python
crim_ptratio_bounded = df[df["CHAS"]==1].filter(["CRIM", "PTRATIO"])
```

```python
plt.scatter(crim_ptratio_bounded["CRIM"], crim_ptratio_bounded["PTRATIO"])
plt.xlabel("Crime Rate"), plt.ylabel("PTRATIO"), plt.title("Crime rate versus Pupile Teacher Ratio, bounded by river.")
plt.show()
```

```python
crim_ptratio_unbounded = df[df["CHAS"]==0].filter(["CRIM","PTRATIO"])
```

```python
plt.scatter(crim_ptratio_unbounded["CRIM"], crim_ptratio_unbounded["PTRATIO"])
plt.xlabel("CRIM"), plt.ylabel("PTRATIO"), plt.title("Crime rate versus Pupil Teacher Ratio, unbounded by river.")
plt.show()
```

```python
b_vs_dis = df.filter(["B","DIS"])
plt.scatter(b_vs_dis.B, b_vs_dis.DIS)
plt.xlabel("Proportion of Black People"), plt.ylabel("Weighted Distance to Employment Center")
plt.title("Proportion of Black People vs. Weighted Distance to Employment Center")
plt.show()
```

```python
medv_vs_nox = df.filter(["MEDV","NOX"])
medv_vs_indus = df.filter(["MEDV", "INDUS"])
plt.scatter(medv_vs_nox.MEDV, medv_vs_nox.NOX)
plt.scatter(medv_vs_indus.MEDV, medv_vs_indus.INDUS)
plt.xlabel("Median Value of Homes")
plt.ylabel("Nitric Oxides Concentration, Proportion of Non-Retail Business")
plt.title("Median Value of Homes compared to Nitric Oxides concentration and Non-Retail Business Proportion")
plt.show()
```
