```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Dataset: City of Seattle Expenditures 2012 - 2014
This Data sets describes actual expenditures for 2012, 2013 adopted budgetary expenditures, and 2014 endorsed and proposed budget levels.

Data is also broken down by department.

Step 1: Bring in the data

```python
data = pd.read_csv('Expenditures-dollars.csv')
df = pd.DataFrame(data).fillna(0)
```

```python
df
```

Question 1: What is the total spending for each department in 2012?

```python
departments = df.Department.unique()
res_list = []
for department in departments:
    dep_spending = df[df['Department'] == department].filter(items=["2012 Actual"])
    res_list.append((department, dep_spending.sum()))
res_list    
```

Question 2: Change in spending per department (except for departments without data 0.0 values) from 2012 to 2014 Proposed

```python
dep_2012 = []
dep_2014 = []
for department in departments:
    dep_2012.append((department, df[(df["Department"] == department) & (df["2012 Actual"] != 0.0)].filter(items=["2012 Actual"]).sum()))
    dep_2014.append((department, df[(df["Department"] == department) & (df["2012 Actual"] != 0.0)].filter(items=["2014 Proposed"]).sum()))
res_list = []
for i in range (len(dep_2012)):
    res_list.append((dep_2012[0], dep_2014[i][1] - dep_2012[i][1]))
res_list
```

Question 3: Plot spending for each department for each year.

```python
df2 = df.filter(items=["Department", "2012 Actual", "2013 Adopted", "2014 Endorsed", "2014 Proposed"])
df2.groupby(df2["Department"]).sum()

df2.plot(title="Department Spending")
plt.show()
```

Question 4: Total spent each year

```python
years = ["2012 Actual", "2013 Adopted", "2014 Endorsed", "2014 Proposed"]
for year in years:
    print(year, df[year].sum())
```
